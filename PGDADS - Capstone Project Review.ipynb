## Handling Missing Values

On this section, we probably need to find out what is the magnitude of missing values in the dataset and what treatments do we want to apply to overcome that. For now, let's take alook at which features do we need to address and how.

```python
total_obs = df.isnull().count()
calculate_missing = df.isnull().sum()
total_missing = calculate_missing[calculate_missing.values>0].sort_values(ascending=False)
perc_missing = round(total_missing/total_obs*100,2)
perc_missing = perc_missing[perc_missing.values>0].sort_values(ascending=False)
col_missing_data = pd.concat([total_missing,perc_missing],axis=1,keys=['#','%']).sort_values('#',ascending=False)

summary_table = tabulate(col_missing_data, headers=["Features", "No of Obs Missing #", "No of Obs Missing %"])
print('There are {} columns with missing data.'.format(col_missing_data.shape[0]))
print('These columns are as follows:\n')
print(summary_table)
```

There are features with high proportion of missing data</b>, with <b>PoolQC</b> being the top. For a simpler or straightforward approach, it would be easier to remove these rows in total and not comprising on the overall purpose of this exercise. Alternatively, we can impute these missing values with median/mode.

###  Remove features with high proportion of missing data

Filter those features with more than 40% of values missing. As a general rule of thumb, we probably remove them in totality.

```python
col_missing_data[col_missing_data['%'] >= 40]
```

```python
missing=col_missing_data.reset_index()
missing_high=missing['index'][missing['%']>=40]
list_to_remove = []
for i,v in enumerate(missing_high):
    list_to_remove.append(v)
df_after_fix1 = df.drop(list_to_remove, axis=1)
```

```python
df_after_fix1.describe().T
```

```python
print('These 5 features are removed: {}.'.format(list_to_remove))
```

After excluding all missing data, the new dataset consists of:

```python
print('After excluded these features, there are {} observations and {} features.'.format(*df_after_fix1.shape))
print('\t Of which, {} numeric features'.format(df_after_fix1.select_dtypes(exclude='object').shape[1]))
print('\t And, {} non-numeric/object features.'.format(df_after_fix1.select_dtypes(include='object').shape[1]))
```

You probably have noticed <b>80</b> features are now left with <b>75</b>. And for the rest of features with some missing values, we will impute using median or mode depending on the features

### Impute data with median or mode

Whilst for those features with < 40% missing data, we will fill them up. This part requires some level of data imputation (i.e. applying median or mode depending on the data type).

```python
col_missing_data[col_missing_data['%'] < 40]
```

```python
missing_low=missing['index'][missing['%']<40]
list_to_impute = []
for i,v in enumerate(missing_low):
    list_to_impute.append(v)
df_need_fixes = df_after_fix1[list_to_impute].copy()
print('After removed 5 features, we still need to review the remaining {} features.'.format(df_need_fixes.shape[1]))
print('\t Of which, {} numeric features'.format(df_need_fixes.select_dtypes(exclude='object').shape[1]))
print('\t And, {} categorical features.'.format(df_need_fixes.select_dtypes(include='object').shape[1]))
```

```python
# this copied df will be used for data imputation process
df_need_fixes.head(15).T
```

Looks like it, we might need to handle categorical and numeric features separately. For a start, let's take alook at the categorical features.

```python
#list of categorical features and apply MODE function to missing values.
apply_mode = df_need_fixes.select_dtypes(include='object').mode()
apply_mode
```

```python
#copy all the columns into a new df
df_after_fix2 = df_after_fix1.copy()
```

```python
#create new columns into the new df
for col in apply_mode:
    df_after_fix2['new_'+str(col)] = df_after_fix1[col].copy()
    df_after_fix2['new_'+str(col)] = df_need_fixes[col].fillna(apply_mode[col][0])
```

```python
#remove all old columns - this will give us the same set of columns
df_after_fix2 = df_after_fix2.drop(apply_mode.columns, axis=1)
```

```python
basic_stats['Q1']=df_need_fixes.select_dtypes(exclude='object').describe().T['25%'] #lower quartile
basic_stats['Q3']=df_need_fixes.select_dtypes(exclude='object').describe().T['75%'] #upper quartile
basic_stats['IQR']=basic_stats['Q3']-basic_stats['Q1']
basic_stats
```

## Multicollinearity

In regression, "multicollinearity" refers to predictors that are correlated with other predictors.  Multicollinearity occurs when your model includes multiple factors that are correlated not just to your response variable, but also to each other. In other words, it results when you have factors that are a bit redundant.

Unfortunately, there is a bit of mismatch in the distribution above. We have one variable as categorical and another as continuous data.

What we need is something that will look like correlation, but will work with categorical values — or more formally, we’re looking for a measure of association between two categorical features. Introducing: Cramér’s V. It is based on a nominal variation of Pearson’s Chi-Square Test.

```python
def cramers_v(x, y):
    from scipy.stats import chi2_contingency
    confusion_matrix = pd.crosstab(x,y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return round(np.sqrt(phi2corr/min((kcorr-1),(rcorr-1))),4)
```

```python
cramers_v(new_df['GarageCars'],new_df['GarageArea'])
```

## Managing Outliers

Come to the last part of data construction, managing outliers, I find boxplot does a really good job and applying Z-score (if you know how to do that). 

I have written a function below to identify outliers.

```python
#for outliers detection
def detect_outliers(df):
    stats_for_outliers = pd.concat([pd.DataFrame(df.select_dtypes(exclude='object').describe().T),\
                         pd.DataFrame(df.select_dtypes(exclude='object').median(),columns=['median'])],axis=1)
    stats_for_outliers['Q1']=df.select_dtypes(exclude='object').describe().T['25%'] #lower quartile
    stats_for_outliers['Q3']=df.select_dtypes(exclude='object').describe().T['75%'] #upper quartile
    stats_for_outliers['IQR']=stats_for_outliers['Q3']-stats_for_outliers['Q1']
    stats_for_outliers['low_xtrem']=np.where(stats_for_outliers['Q1']-(stats_for_outliers['IQR']*1.5)<stats_for_outliers['min'],stats_for_outliers['min'],stats_for_outliers['Q1']-(stats_for_outliers['IQR']*1.5))
    stats_for_outliers['hi_xtrem']=np.where(stats_for_outliers['Q3']+(stats_for_outliers['IQR']*1.5)>stats_for_outliers['max'],stats_for_outliers['max'],stats_for_outliers['Q3']+(stats_for_outliers['IQR']*1.5))
    #del stats_for_outliers['count']
    return stats_for_outliers
```

```python
check = detect_outliers(clean_df.drop('SalePrice',axis=1))
check['outlier_ind'] = np.where(check['max'] - check['hi_xtrem']>0,'Y','N')
outliers=['count','min','max','Q1','Q3','IQR','low_xtrem','hi_xtrem','outlier_ind']
check[outliers].T
```

On normal circumstances, a simple linear regression may work as effective as any other algorithm. But, if we have large number of features like in this case, and test score is relatively poor than the training score then it’s the problem of over-generalization or over-fitting.

Ridge and Lasso regression are some of the simple techniques to reduce model complexity and prevent over-fitting which may result from simple linear regression.

## Ridge Regression (Ridge)

In ridge regression, the cost function is altered by adding a penalty equivalent to square of the magnitude of the coefficients.

So ridge regression puts constraint on the coefficients (w). The penalty term (lambda) regularizes the coefficients such that if the coefficients take large values the optimization function is penalized. So, ridge regression shrinks the coefficients and it helps to reduce the model complexity and multi-collinearity. 

This is equivalent to saying minimizing the cost function as like linear regression. So lower the constraint (low λ) on the features, the model will resemble linear regression model.

## Kernel Ridge Regression

## Non-Prob Binary | Support Vector Regression

Support Vector Machine can also be used as a regression method, maintaining all the main features that characterize the algorithm (maximal margin). The Support Vector Regression (SVR) uses the same principles as the SVM for classification, with only a few minor differences. First of all, because output is a real number it becomes very difficult to predict the information at hand, which has infinite possibilities. In the case of regression, a margin of tolerance (epsilon) is set in approximation to the SVM which would have already requested from the problem. But besides this fact, there is also a more complicated reason, the algorithm is more complicated therefore to be taken in consideration. However, the main idea is always the same: to minimize error, individualizing the hyperplane which maximizes the margin, keeping in mind that part of the error is tolerated.
- <b>Rule</b>: Standardise all independent variables & target variable before deploying SVR

```python
def svr(X_train, X_test, y_train, y_test):
    #Feature Scaling is required for SVR
    sc_X = StandardScaler()
    X_train_svm = sc_X.fit_transform(X_train)
    X_test_svm = sc_X.fit_transform(X_test)
    svr = SVR(random_state=24)
    # Construct the hyperparameter grid
    param_grid = {'kernel': ['rbf','linear', 'poly', 'rbf', 'sigmoid', 'precomputed'],
                  'C': [1e3], 
                  'epsilon': [0.001],
                 }
    # Instantiate the GridSearchCV object using the estimator object,
    # the hyperparameter grid, & 5-fold cross-validation
    grid_svr = GridSearchCV(svr, param_grid=param_grid, cv=5)
    grid_svr.fit(X_train_svm, y_train)
    y_pred = grid_svr.predict(X_test_svm)
    print('R2 Score   :', round(r2_score(y_validation, y_pred),4))
    print('MSE        :', round(metrics.mean_squared_error(y_validation, y_pred),4))
    print('RMSE       :', round(np.sqrt(metrics.mean_squared_error(y_validation, y_pred)),4))
    print('Optimal hyperparameter(s): {}.'.format(dict(grid_svr.best_params_)))
    print('Optimal Estimator:\n{}'.format(grid_svr.best_estimator_))
```

## Boosting Model | Gradient Boost

## Recursive Feature Elimination (RFE)

For a start, we test out how RFE works out in any algorithm model. To do that, we can use recursive feature elimination proces. Recursive Feature Elimination (RFE) method works by recursively removing attributes and building a model on those attributes that remain. It uses accuracy metric to rank the feature according to their importance. The RFE method takes the model to be used and the number of required features as input. 

```python
x = xgb.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0,
             importance_type='gain', learning_rate=0.02, max_delta_step=0,
             max_depth=5, min_child_weight=1, missing=None, n_estimators=1000,
             n_jobs=1, nthread=None, objective='reg:linear', random_state=24,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
             silent=None, subsample=1, verbosity=1)
#total number of features for simulation
nof_list=np.arange(15,60)      

high_score=0
#Variable to store the optimum no of features
nof=0           
score_list =[]
for n in range(len(nof_list)):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 24)
    rfe = RFE(x,nof_list[n])
    rfe.fit(X_train,y_train)
    score = r2_score(rfe.predict(X_validation), y_validation)
    score_list.append(score)
    if(score > high_score):
        high_score = score
        nof = nof_list[n]
print("Optimal number of features : %d" % nof)
print("Total max number of features trialling : %d" % rfe.n_features_)
```

## Stacking Model vs Mixed Model

Stacking is the process of using different Machine Learning models one after another, where you add the predictions from each model to make a new feature. Model stacking should always be accompanied by cross validation, to reduce overfitting models to training data. 

Model stacking will seem like a simple technique to improving your results, when you understand what happens inside the algorithm. Though, there is many components interacting, and keeping track of all of them can be quite difficult, especially when first learning this concept.

Stacking machine learning models is done in layers, and there can be arbitrarily many layers, dependent on exactly how many models you have trained, along with the best combination of these models. For an instance, the first layer could boil down to learning some feature that has great predictive power, while the next layer might be doing something else, like reducing noise.

We put models stacks in layers, and usually with a different purpose. At the end, we get a final dataset, that we feed into a last model. The last model is called a meta-learner (e.g. meta-regressor or meta-classifier), and its purpose is to generalize all the features from each layer into the final predictions.

Apparently, this is a very famous model in Kaggle competition.
